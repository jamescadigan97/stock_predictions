```python
#Dependencies
from tensorflow.keras.layers import LSTM
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Activation
from numpy import array
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
#Load Data
#VMW = pd.read_csv("VMW.csv")
#VMW = pd.read_csv("VMW2.csv")
#VMW
CSCO = pd.read_csv("CSCO.csv")
CSCO
```

```python
del CSCO["Open"]
del CSCO["Close"]
del CSCO["High"]
del CSCO["Low"]
del CSCO["Adj Close"]
```

```python
CSCO["Next Volume"]=CSCO["Volume"].shift(-1)
CSCO=CSCO.dropna(how='any')
CSCO
```

```python
#Shape Data
CSCO_vol = CSCO['Volume'].to_numpy()
CSCO_vol = np.float32(CSCO_vol)
X = CSCO_vol.reshape(-1,1)
print("Volume shape")
print(np.shape(X))
CSCO_next = CSCO['Next Volume'].to_numpy()
CSCO_next = np.float32(CSCO_next)
y = CSCO_next.reshape(-1,1)
print("Next Volume shape")
print(np.shape(y))
```

```python
X = X.reshape(1,754,1)
y = y.reshape(1,754,1)
```

```python
#Define Model
model = Sequential()
model.add(LSTM(32))
model.add(Dense(1))
#model.add(Activation('sigmod'))
```

```python
#Compile model
model.compile(optimizer='sgd', loss='mean_squared_error', metrics=['mean_squared_error'])
```

```python
#Fit model
#model.fit(X, y, batch_size=10, epoch=100)
model.fit(
    X,
    y,
    batch_size=32,
    epochs=50,
    shuffle=True,
    verbose=2
)
```

```python
#Evaluate model
loss, accuracy = model.evaluate(X, y)
```

```python
#Make predictions
predictions = model.predict(X)
```
