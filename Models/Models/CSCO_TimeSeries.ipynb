```python
#Dependencies
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
from tensorflow.keras.layers import LSTM
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Activation
from tensorflow.keras.layers import Dropout
from numpy import array
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
CSCO = pd.read_csv("CSCO.csv", date_parser=True)
CSCO
```

```python
#Change paramters
cut_off_date = "2019-1-1"
x_train_days = 120
num_epochs = 50
num_batch = 32
data_dim = 1
units1=50
units2=50
```

```python
CSCO_train_ = CSCO[CSCO["Date"]>cut_off_date].copy()
CSCO_train_
```

```python
train_CSCO=CSCO_train_.drop(["Date","Adj Close","Volume","High", "Low", "Close"], axis = 1)
train_CSCO=train_CSCO.tail(30)
```

```python
from sklearn.preprocessing import MinMaxScaler
scalar = MinMaxScaler()
CSCO_train = scalar.fit_transform(train_CSCO)
CSCO_train
```

```python
model = Sequential()

model.add(LSTM(units=units1, activation="relu", return_sequences= True))
model.add(Dropout(0.2))

#model.add(LSTM(units=50, activation="relu", return_sequences= True))
#model.add(Dropout(0.2))

#model.add(LSTM(units=50, activation="relu", return_sequences= True))
#model.add(Dropout(0.2))

model.add(LSTM(units=units2, activation="relu"))
model.add(Dropout(0.2))

model.add(Dense(1))

model.compile(optimizer='adam', loss='mse')
```

```python
time_train=CSCO_train
```

```python
n_input = 29
n_features = 1
generator = TimeseriesGenerator(time_train, time_train, length=n_input, batch_size=6)
```

```python
model.fit_generator(generator, steps_per_epoch=1, epochs=200, verbose=1)
#model.fit_generator(generator,epochs=10,verbose=1)
```

```python
pred_list = []
batch = time_train[-n_input:].reshape((1, n_input, n_features))
for i in range(12):   
    pred_list.append(model.predict(batch)[0]) 
    batch = np.append(batch[:,1:,:],[[pred_list[i]]],axis=1)
```
